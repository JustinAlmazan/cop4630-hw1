```python
# Read and plot data
data = pd.read_csv('data_nonlinear.csv')
x = data.iloc[:, 0]
y = data.iloc[:, 1]
plt.scatter(x, y)
plt.show()
```

```python
# Initialize coefficients to 0
a = 0 # 3rd degree
b = 0 # 2nd degree
c = 0 # 1st degree
d = 0 # 0th degree

# Givens (learning rate & number of epochs)
LR = 1e-6
numSteps = int(1e4)
n = float(len(x))

for i in range(numSteps):

    # Calculate predicted value for Y using cubic model
    y_p = a*(x**3) + b*(x**2) + c*x + d
    
    # Derivatives for gradient descent
    d_a = (-2/n) * sum((x**3)*(y-y_p))
    d_b = (-2/n) * sum((x**2)*(y-y_p))
    d_c = (-2/n) * sum((x**1)*(y-y_p))
    d_d = (-2/n) * sum((x**0)*(y-y_p))
    
    # Adjustment
    a = a - LR * d_a
    b = b - LR * d_b
    c = c - LR * d_c
    d = d - LR * d_d

# Coefficients for final cubic model
print (a, b, c, d)
```

```python
plt.scatter(x, y)

# Calculate and plot predicted Y value for all X
for i in range(len(x)):
    
    y_p = a*(x**3) + b*(x**2) + c*x + d
    plt.plot(x, y_p, 'o', color='red')

plt.show()
```

```python
# Read new data
from mpl_toolkits.mplot3d import Axes3D
data = pd.read_csv('data_two_variables.csv')
x1 = data.iloc[:, 0]
x2 = data.iloc[:, 1]
y = data.iloc[:, 2]

# Make plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot points & show
ax.scatter(x1, x2, y, c='b', marker='o')
plt.show()
```

```python
# Coefficients for linear model
m1 = 0
m2 = 0
m3 = 0
LR = 1e-3
numSteps = int(1e4)
n = float(len(y))

# Same logic as part 3
for i in range(numSteps):
    y_p = m1*x1 + m2*x2 + m3

    d_m1 = (-2/n) * sum(x1*(y-y_p))
    d_m2 = (-2/n) * sum(x2*(y-y_p))
    d_m3 = (-2/n) * sum(y-y_p)

    m1 = m1 - LR * d_m1
    m2 = m2 - LR * d_m2
    m3 = m3 - LR * d_m3

print (m1, m2, m3)
    
```

```python
# Make plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot points
ax.scatter(x1, x2, y, c='b', marker='o')

# Calculate precicted value for y using linear model
for i in range(len(x1)):
    y_p = m1*x1 + m2*x2 + m3
    plt.plot(x1, x2, y_p, 'o', c='red')

plt.show()
```
